```python
import pandas as pd

df = pd.read_csv("census_data.csv")

mod_df=df.copy()

mod_df.drop('Unnamed: 7', axis=1, inplace=True)
mod_df.rename(columns={'[["A_HGA"':'AttainedEd', 'PRDTRACE':'Race', 'HTOTVAL':'TotalWages', 'PEHSPNON':'HispanicInd', 'A_SEX':'Sex', 'A_MARITL':'MaritalStatus','state]':'State'}, inplace=True)
mod_df.head()
```

```python
mod_df['AttainedEd'] = mod_df['AttainedEd'].str.replace("[\"", "")
mod_df['AttainedEd'] = mod_df['AttainedEd'].str.replace("\"", "")
mod_df['State'] = mod_df['State'].str.replace("]", "")
mod_df.head()
```
